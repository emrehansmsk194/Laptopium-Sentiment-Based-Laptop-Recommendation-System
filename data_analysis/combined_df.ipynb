```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("combined_products.csv")
df
```

```python
df.info()
```

```python
new_df = df.copy()
new_df
```

```python
def fix_price(price):
    # Eƒüer fiyat bir string ve ',' i√ßeriyorsa
    if isinstance(price, str) and ',' in price:
        # Noktadan √∂nceki kƒ±smƒ± al (binlik ayƒ±rƒ±cƒ±larƒ± kaldƒ±r)
        return price.split(',')[0] + ' TL'
    return price  # Eƒüer fiyat d√ºzg√ºnse dokunma

new_df['Price'] = new_df['Price'].apply(fix_price)
```

```python
new_df.info()
```

```python
new_df
```

```python
new_df = new_df.sample(frac=1).reset_index(drop=True)
new_df.info()
```

```python
new_df
```

```python
new_df = new_df.dropna(subset=['Price'])
new_df.isna().sum()
```

```python
new_df = new_df.dropna(subset=['Reviews'])
new_df.isna().sum()
```
