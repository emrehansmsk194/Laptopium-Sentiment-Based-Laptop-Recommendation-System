```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Selenium WebDriver ayarları
chrome_driver_path = "C:\\Users\\EMREHAN\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"
chrome_options = Options() 
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")

# WebDriver'ı başlat
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)
wait = WebDriverWait(driver, 15)  # 15 saniyeye kadar bekle

def extract_product_details(url):
    try:
        driver.get(url)

        # Ürün adını çek
        try:
            product_name_element = wait.until(
                EC.presence_of_element_located(
                    (By.XPATH, "//h1[@data-test-id='title']")
                )
            )
            product_name = product_name_element.text.strip()
        except Exception as e:
            product_name = None
            print(f"Ürün adı çekilirken hata: {e}")

        # Fiyatı çek
        try:
            price_element = wait.until(
                EC.presence_of_element_located(
                    (By.XPATH, "//div[@data-test-id='default-price']//span")
                )
            )
            price = price_element.text.strip()
        except Exception as e:
            price = None
            print(f"Fiyat çekilirken hata: {e}")

        # Resim URL'sini çek (Placeholder olmayanı al)
        try:
            image_elements = driver.find_elements(By.XPATH, "//picture//img")
            image_url = None
            for img in image_elements:
                src = img.get_attribute("src")
                # Placeholder resimleri filtrele
                if "hepsiburada.net" in src and "banners" not in src:
                    image_url = src
                    break
        except Exception as e:
            image_url = None
            print(f"Resim URL çekilirken hata: {e}")

        print("Yeni Ürün Adı :", product_name)
        print("Yeni Fiyat    :", price)
        print("Yeni Resim URL:", image_url)
        print("-" * 50)

        return product_name, price, image_url

    except Exception as e:
        print(f"Genel hata: {e}")
        return None, None, None


# İlk 5 URL üzerinde döngü
for index, row in df.iterrows():
    product_url = row["Link"]
    new_product_name, new_price, new_image_url = extract_product_details(product_url)

    if new_product_name:
        df.at[index, "Product Name"] = new_product_name
    if new_price:
        df.at[index, "Price"] = new_price
    if new_image_url:
        df.at[index, "Image URL"] = new_image_url

# İşiniz bittikten sonra WebDriver'ı kapatın.
driver.quit()
```
