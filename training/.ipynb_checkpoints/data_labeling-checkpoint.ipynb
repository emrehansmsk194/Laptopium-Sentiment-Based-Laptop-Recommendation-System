```python
import pandas as pd
import re

# CSV dosyasını yükleyin
df = pd.read_csv("all_laptop_reviews.csv")

# Yorumlar dosyasını tüm içeriğiyle okuyun
with open("C://Users/Emrehan/Desktop/yorumlar.txt", encoding="utf-8") as f:
    text = f.read()

# Dosyayı "Yorum [numara]:" başlıklarıyla bloklara ayırın
blocks = re.split(r"(?m)^Yorum\s+\d+:", text)

# İlk blok boş olabilir; sonraki bloklardan her biri bir yoruma ait
valid_answers = []
for block in blocks[1:]:
    # Her blokta, "Bu yorumu nasıl etiketlemek istersiniz? (pozitif/negatif/nötr):" ifadesinden sonra gelen tüm
    # geçerli cevapları (pozitif, negatif veya nötr) buluyoruz.
    answers = re.findall(
        r"Bu yorumu nasıl etiketlemek istersiniz\?\s*\(pozitif/negatif/nötr\):\s*(pozitif|negatif|nötr)",
        block,
        flags=re.IGNORECASE
    )
    if answers:
        # Eğer birden fazla cevap varsa, sonuncusunu alıyoruz
        valid_answers.append(answers[-1].strip().lower())
    else:
        # Eğer hiçbir geçerli cevap bulunamazsa, hata vermek yerine varsayılan "nötr" ekleyebilirsiniz
        valid_answers.append("nötr")

print("Toplam geçerli etiket sayısı:", len(valid_answers))  # Beklenen: 1300

# CSV dosyasındaki yorumlardan, etiket sayısı kadarını seçip yeni 'label' sütunu olarak ekliyoruz
df_subset = df.head(len(valid_answers))
df_subset["label"] = valid_answers

# Sonuçları yeni bir CSV dosyasına kaydediyoruz
df_subset.to_csv("labeled_reviews.csv", index=False)
```
