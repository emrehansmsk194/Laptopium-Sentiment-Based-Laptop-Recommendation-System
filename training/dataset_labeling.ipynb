```python
def clean_review(review):
    if isinstance(review, str):
        # 'Comment:' kısmından böl ve sadece yorum kısmını al
        if '| Comment:' in review:
            review = review.split('| Comment:')[1].strip()
        # 'Rating:' kısmını kaldır
        if 'Rating:' in review:
            review = review.split('Rating:')[0].strip()
        # Başına '|' ekle
        return '| ' + review
    return review

# Review sütununu temizle
if 'Reviews' in deneme_df2.columns:
    deneme_df2['Reviews'] = deneme_df2['Reviews'].apply(clean_review)
elif 'Review' in deneme_df2.columns:
    deneme_df2['Review'] = deneme_df2['Review'].apply(clean_review)
```

```python
df = pd.read_csv("C://Users/Emrehan/Desktop/Emrehan Simsek/veri çekme selenium/data_analysis/cleaned_products.csv")
```

```python
review_col = 'Reviews' if 'Reviews' in df.columns else 'Review' if 'Review' in df.columns else None

if review_col:
    def clean_review(review):
        if isinstance(review, str):
            # Comment: kısmını kontrol et ve sil
            if 'Comment:' in review:
                review = review.split('Comment:')[1].strip()
            # Rating: kısmını kontrol et ve sil
            if 'Rating:' in review:
                review = review.split('Rating:')[0].strip()
            # Başına sadece '|' ekle
            return '| ' + review
        return review

    # Temizleme işlemini uygula
    df[review_col] = df[review_col].apply(clean_review)
```

```python
split_symbols = r'\|\|?\s*|;\s*'

# Her satırdaki yorumları ayırıp, her birini ayrı satır yapalım
df_exploded = df.assign(Reviews=df['Reviews'].str.split(split_symbols)).explode('Reviews')

# Boşlukları temizleyelim ve boş yorumları silelim
df_exploded['Reviews'] = df_exploded['Reviews'].str.strip()
df_exploded = df_exploded[df_exploded['Reviews'] != '']
```
