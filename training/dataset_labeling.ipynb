```python
def clean_review(review):
    if isinstance(review, str):
        # 'Comment:' kÄ±smÄ±ndan bÃ¶l ve sadece yorum kÄ±smÄ±nÄ± al
        if '| Comment:' in review:
            review = review.split('| Comment:')[1].strip()
        # 'Rating:' kÄ±smÄ±nÄ± kaldÄ±r
        if 'Rating:' in review:
            review = review.split('Rating:')[0].strip()
        # BaÅŸÄ±na '|' ekle
        return '| ' + review
    return review

# Review sÃ¼tununu temizle
if 'Reviews' in deneme_df2.columns:
    deneme_df2['Reviews'] = deneme_df2['Reviews'].apply(clean_review)
elif 'Review' in deneme_df2.columns:
    deneme_df2['Review'] = deneme_df2['Review'].apply(clean_review)
```

```python
df = pd.read_csv("C://Users/Emrehan/Desktop/Emrehan Simsek/veri Ã§ekme selenium/data_analysis/cleaned_products.csv")
```

```python
review_col = 'Reviews' if 'Reviews' in df.columns else 'Review' if 'Review' in df.columns else None

if review_col:
    def clean_review(review):
        if isinstance(review, str):
            # Comment: kÄ±smÄ±nÄ± kontrol et ve sil
            if 'Comment:' in review:
                review = review.split('Comment:')[1].strip()
            # Rating: kÄ±smÄ±nÄ± kontrol et ve sil
            if 'Rating:' in review:
                review = review.split('Rating:')[0].strip()
            # BaÅŸÄ±na sadece '|' ekle
            return '| ' + review
        return review

    # Temizleme iÅŸlemini uygula
    df[review_col] = df[review_col].apply(clean_review)
```
